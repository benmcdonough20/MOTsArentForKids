# Easy data processing
Using the `Experiment` and `DataRun` class to generalize processing data. General usage:

1. Create an `Experiment`
2. Use the `structured_data` method to extract the parameter of interest
3. Plot

```python
import datarun
from matplotlib import pyplot as plt
import numpy as np
```

```python
#If you make a change to datarun.py, run this cell or restart your kernel
import importlib
importlib.reload(datarun)
```

```python
d = datarun.Experiment(
    vstart = 2.5,
    vend = 3.1,
    vstep = .15, 
    idx_start = 16,
    datapath = "/home/ben/Documents/Repos/MOTsArentForKids/data/mar09/MOT_current/",
    numtrials = 1,
    blob_dim = 400, #size of blob box to fit
    box = 3, #size of box for image median filter
    mask_box = 50, #size of box for image mask
    circle = [(530, 675), 300], #center/radius of viewport circle
    avg_area = (200, 100, 300, 200), #area for capturing background noise
    mask = .15
)
```

```python
d.data[2].plot_fit()
```

```python
N_list = d.structure_data(lambda d: d.atom_number())
plt.plot(N_list.keys(), [np.mean(n) for n in N_list.values()], marker = "x")
```
